## 2 概述、线性代数和 NDArray

### 2.1 理论计算题

已知 $\mathbf{a} = [2, -1, 3]^\top$，$\mathbf{b} = [1, 4, -2]^\top$，

$$A = \begin{bmatrix} 1 & 0 & 2 \\\\ -1 & 3 & 1 \end{bmatrix}, \quad B = \begin{bmatrix} 2 & 1 \\\\ 0 & -1 \\\\ 3 & 2 \end{bmatrix}$$

**1. 向量点积 $\mathbf{a} \cdot \mathbf{b}$：**

$$\mathbf{a} \cdot \mathbf{b} = 2 \times 1 + (-1) \times 4 + 3 \times (-2) = 2 - 4 - 6 = -8$$

**2. 矩阵乘法 $A \times B$：**

$A$ 的形状为 $2 \times 3$，$B$ 的形状为 $3 \times 2$，结果矩阵形状为 $\mathbf{2 \times 2}$。

$$A \times B = \begin{bmatrix} 1\cdot2+0\cdot0+2\cdot3 & 1\cdot1+0\cdot(-1)+2\cdot2 \\\\ (-1)\cdot2+3\cdot0+1\cdot3 & (-1)\cdot1+3\cdot(-1)+1\cdot2 \end{bmatrix} = \begin{bmatrix} 8 & 5 \\\\ 1 & -2 \end{bmatrix}$$

**3. 向量 $\mathbf{a}$ 的 Frobenius 范数：**

$$\|\mathbf{a}\|_F = \sqrt{2^2 + (-1)^2 + 3^2} = \sqrt{4 + 1 + 9} = \sqrt{14} \approx 3.742$$

### 2.2 编程题

```python
import numpy as np

# 1. 创建形状为 3x4 的随机矩阵 X，元素服从标准正态分布
np.random.seed(42)
X = np.random.randn(3, 4)
print('X (3x4 随机矩阵，标准正态分布):')
print(X)

# 2. 创建形状为 4x2 的全 1 矩阵 Y
Y = np.ones((4, 2))
print('\nY (4x2 全1矩阵):')
print(Y)

# 3. 计算矩阵乘法 Z = X x Y
Z = X @ Y
print('\nZ = X x Y (3x2 矩阵):')
print(Z)

# 4. 输出 Z 的第一行和第二列交叉处的元素，以及 Z 的第 2 行所有元素
print('\nZ[0, 1] (第一行、第二列):', Z[0, 1])
print('Z[1, :] (第 2 行所有元素):', Z[1, :])

# 5. 计算 Z 的 Frobenius 范数
frob_norm = np.linalg.norm(Z, 'fro')
print('\nZ 的 Frobenius 范数:', frob_norm)
```

## 3 概率与统计

### 3.1 理论计算题

设事件 $D$ 为患病，$T^+$ 为检测阳性。已知：

- 先验概率：$P(D) = 0.001$，$P(\\bar{D}) = 0.999$
- 灵敏度：$P(T^+ \\mid D) = 0.99$
- 假阳性率：$P(T^+ \\mid \\bar{D}) = 0.02$

由全概率公式：

$$P(T^+) = 0.99 \\times 0.001 + 0.02 \\times 0.999 = 0.00099 + 0.01998 = 0.02097$$

由贝叶斯公式：

$$P(D \\mid T^+) = \\frac{0.00099}{0.02097} \\approx 0.04721 \\approx 4.72\\%$$

**结论**：即使检测为阳性，真正患病的概率约为 **4.72%**，体现了低患病率场景下假阳性问题的严重性。

### 3.2 编程题：中心极限定理模拟

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(0)
n = 10000  # 每次抽样的变量数
m = 1000   # 重复次数

# 1. 生成 m 次均值
samples = np.random.uniform(0, 1, size=(m, n))
means = samples.mean(axis=1)

# 理论参数：U(0,1) 均值0.5，样本均值方差 1/(12*n)
mu_theory = 0.5
sigma_theory = np.sqrt(1 / (12 * n))

# 2. 绘制直方图 + 理论正态分布曲线
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(means, bins=40, density=True, alpha=0.6, color='steelblue', label='样本均值直方图')
x = np.linspace(means.min(), means.max(), 300)
ax.plot(x, norm.pdf(x, mu_theory, sigma_theory), 'r-', lw=2,
        label=f'理论正态分布 N({mu_theory}, {sigma_theory**2:.2e})')
ax.set_xlabel('样本均值')
ax.set_ylabel('概率密度')
ax.set_title('中心极限定理验证：U(0,1) 样本均值分布 (n=10000, m=1000)')
ax.legend()
plt.tight_layout()
plt.show()

# 3. 计算实际方差
actual_variance = np.var(means)
print(f'样本均值的实际方差: {actual_variance:.6e}')
print(f'理论方差 1/(12n):   {1/(12*n):.6e}')
```

## 4 导数、反向传播和复杂度

### 4.1 理论计算题

给定 $z = (w_1 x_1 + w_2 x_2 - y)^2$，其中 $x_1=2, x_2=1, y=3$。

令 $u = w_1 x_1 + w_2 x_2 - y$，则 $z = u^2$。

**1. 偏导数表达式：**

$$\\frac{\\partial z}{\\partial w_1} = 2(w_1 x_1 + w_2 x_2 - y) \\cdot x_1$$

$$\\frac{\\partial z}{\\partial w_2} = 2(w_1 x_1 + w_2 x_2 - y) \\cdot x_2$$

**2. 代入 $w_1=0.5, w_2=1$：**

$$u = 0.5 \\times 2 + 1 \\times 1 - 3 = -1$$

$$\\frac{\\partial z}{\\partial w_1} = 2 \\times (-1) \\times 2 = -4$$

$$\\frac{\\partial z}{\\partial w_2} = 2 \\times (-1) \\times 1 = -2$$

### 4.2 编程题：手动实现计算图反向传播

```python
import torch

# 给定值
x_val, w1_val, w2_val = 2.0, 1.5, 0.5

# 前向计算：输入 x，a = x*w1，b = a+w2，L = b^2
a = x_val * w1_val
b = a + w2_val
L = b ** 2
print(f'前向计算：a = {a}, b = {b}, L = {L}')

# 手动反向传播（链式法则）
# dL/db = 2b
# dL/dw2 = dL/db * db/dw2 = 2b * 1
# dL/dw1 = dL/db * db/da * da/dw1 = 2b * 1 * x
dL_db = 2 * b
grad_w2_manual = dL_db
grad_w1_manual = dL_db * x_val
print(f'手动梯度: dL/dw1 = {grad_w1_manual}, dL/dw2 = {grad_w2_manual}')

# 自动微分验证（PyTorch）
x  = torch.tensor(x_val)
w1 = torch.tensor(w1_val, requires_grad=True)
w2 = torch.tensor(w2_val, requires_grad=True)
a_t = x * w1
b_t = a_t + w2
L_t = b_t ** 2
L_t.backward()
print(f'PyTorch梯度: dL/dw1 = {w1.grad.item()}, dL/dw2 = {w2.grad.item()}')
print(f'结果一致: {abs(grad_w1_manual - w1.grad.item()) < 1e-5 and abs(grad_w2_manual - w2.grad.item()) < 1e-5}')
```

## 5 线性方法、基础优化和 softmax 回归

### 5.1 理论计算题

线性回归 $\hat{y}_i = wx_i + b$，损失 $L = \\frac{1}{n} \\sum_{i=1}^n (y_i - wx_i - b)^2$

**对 $w$ 求偏导：**

$$\\frac{\\partial L}{\\partial w} = -\\frac{2}{n} \\sum_{i=1}^{n} (y_i - wx_i - b)\\, x_i$$

**对 $b$ 求偏导：**

$$\\frac{\\partial L}{\\partial b} = -\\frac{2}{n} \\sum_{i=1}^{n} (y_i - wx_i - b)$$

### 5.2 编程题：从零实现 softmax 回归（MNIST）

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

# 1. 加载并归一化数据
digits = load_digits()
X_raw = digits.data / 16.0
y_raw = digits.target

# 2. One-hot 编码
num_classes = 10
def one_hot(y, k):
    Y = np.zeros((len(y), k))
    Y[np.arange(len(y)), y] = 1
    return Y

X_train, X_test, y_train, y_test = train_test_split(
    X_raw, y_raw, test_size=0.2, random_state=42)
Y_train = one_hot(y_train, num_classes)

# 3. Softmax 函数和交叉熵损失
def softmax(Z):
    Z_s = Z - Z.max(axis=1, keepdims=True)
    e = np.exp(Z_s)
    return e / e.sum(axis=1, keepdims=True)

def cross_entropy(P, Y):
    return -np.sum(Y * np.log(P + 1e-9)) / len(Y)

# 4. 小批量 SGD 训练（批量大小32，学习率0.1，50 epochs）
W = np.zeros((X_train.shape[1], num_classes))
b = np.zeros((1, num_classes))
lr, epochs, batch_size = 0.1, 50, 32
n_train = len(X_train)

for epoch in range(epochs):
    idx = np.random.permutation(n_train)
    Xs, Ys = X_train[idx], Y_train[idx]
    for i in range(0, n_train, batch_size):
        Xb, Yb = Xs[i:i+batch_size], Ys[i:i+batch_size]
        P = softmax(Xb @ W + b)
        d = (P - Yb) / len(Xb)
        W -= lr * (Xb.T @ d)
        b -= lr * d.sum(axis=0, keepdims=True)
    if (epoch+1) % 10 == 0:
        loss = cross_entropy(softmax(X_train @ W + b), Y_train)
        print(f'Epoch {epoch+1:3d}/50 | 训练损失: {loss:.4f}')

# 5. 测试集准确率
y_pred = softmax(X_test @ W + b).argmax(axis=1)
print(f'\n测试集准确率: {(y_pred == y_test).mean()*100:.2f}%')
```

## 6 最大似然估计和逻辑回归

### 6.1 理论计算题

设 $x_1, \ldots, x_n \overset{\text{i.i.d.}}{\sim} \mathcal{N}(\mu, \sigma^2)$。

**1. 似然函数：**

$$L(\mu, \sigma^2) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\!\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)$$

对数似然：

$$\ell = -\frac{n}{2}\ln(2\pi) - \frac{n}{2}\ln\sigma^2 - \frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2$$

**2. $\mu$ 的 MLE 证明：**

令 $\frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2}\sum(x_i - \mu) = 0$，解得
$\hat{\mu} = \frac{1}{n}\sum_{i=1}^{n} x_i$。$\blacksquare$

**3. $\sigma^2$ 的 MLE 证明：**

令 $v = \sigma^2$，解 $\frac{\partial \ell}{\partial v} = -\frac{n}{2v} + \frac{\sum(x_i-\hat{\mu})^2}{2v^2} = 0$，得
$\hat{\sigma}^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i - \hat{\mu})^2$。$\blacksquare$

### 6.2 编程题：从零实现二分类逻辑回归

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# 1. 生成线性可分的两类数据（各200个样本，特征维度为2）
def gen(n, center):
    return np.random.randn(n, 2) + np.array(center)

X0_tr, X1_tr = gen(200, [-2,-2]), gen(200, [2,2])
X_train = np.vstack([X0_tr, X1_tr])
y_train = np.hstack([np.zeros(200), np.ones(200)])

X0_te, X1_te = gen(50, [-2,-2]), gen(50, [2,2])
X_test = np.vstack([X0_te, X1_te])
y_test = np.hstack([np.zeros(50), np.ones(50)])

# 2. Sigmoid 和二元交叉熵
def sigmoid(z):
    return 1 / (1 + np.exp(-np.clip(z, -500, 500)))

# 3. 梯度下降（学习率0.1，迭代1000次）
w = np.zeros(2)
b = 0.0
for _ in range(1000):
    p = sigmoid(X_train @ w + b)
    d = p - y_train
    w -= 0.1 * (X_train.T @ d) / len(y_train)
    b -= 0.1 * d.mean()

print(f'训练后参数: w = {w}, b = {b:.4f}')

# 4. 绘制决策边界
fig, ax = plt.subplots(figsize=(7, 6))
ax.scatter(X_train[y_train==0,0], X_train[y_train==0,1],
           alpha=0.4, c='steelblue', label='类别 0 (训练)')
ax.scatter(X_train[y_train==1,0], X_train[y_train==1,1],
           alpha=0.4, c='tomato', label='类别 1 (训练)')
x_line = np.linspace(X_train[:,0].min()-1, X_train[:,0].max()+1, 200)
ax.plot(x_line, -(w[0]*x_line + b)/w[1], 'k-', lw=2, label='决策边界')
ax.set_xlabel('特征 1'); ax.set_ylabel('特征 2')
ax.set_title('逻辑回归决策边界'); ax.legend()
plt.tight_layout(); plt.show()

# 5. 测试集准确率
y_pred = (sigmoid(X_test @ w + b) >= 0.5).astype(int)
print(f'测试集准确率: {(y_pred == y_test).mean()*100:.2f}%')
```
